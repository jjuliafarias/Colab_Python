```python
import pandas as pd
```

```python
url = 'https://raw.githubusercontent.com/alura-cursos/pandas-conhecendo-a-biblioteca/main/desafios/alunos.csv'
pd.read_csv(url, sep=';')
```

```python
dados = pd.read_csv(url, sep=',')
dados
```

```python
media = dados['Notas'].mean()
print(f"Média das notas: {media:.2f}")
```

```python
desvio = dados['Notas'].std()

print(f"Desvio padrão das notas: {desvio:.2f}")
```

```python
aluno_maior_nota = dados.loc[dados['Notas'].idxmax()]

print(f"Aluno com a maior nota: {aluno_maior_nota['Nome']} ({aluno_maior_nota['Notas']})")
```
